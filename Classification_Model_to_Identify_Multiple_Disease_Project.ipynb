```python
#Print all of the object data types and their unique values
for column in disease.columns:
    if disease[column].dtype == object:
        print(str(column) + ' : ' + str(disease[column].unique()))
        print(disease[column].value_counts())
        print("_________________________________________________________________")
```

```python
#Remove unneeded columns

#Remove the column EmployeeNumber
disease = disease.drop('prognosis', axis = 1)
```

```python
from sklearn.preprocessing import LabelEncoder

for column in disease.columns:
        if disease[column].dtype == np.number:
            continue
        disease[column] = LabelEncoder().fit_transform(disease[column])
```

```python
# column names
disease.columns
```

```python
# define y
Y = disease.iloc[:, 0].values 
```

```python
# define X
X = disease.iloc[:, 1: disease.shape[1]].values 
```

```python
# split data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
# verify shape
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
# select model
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
```

```python
# train model
model.fit(X_train, Y_train)
```

```python
# predict with model
y_pred=model.predict(X_test)
```

```python
# model evaluation
model.score(X_test,Y_test)
```

```python
# model accuracy
from sklearn.metrics import classification_report
```

```python
# model classification report
print(classification_report(Y_test,y_pred))
```
